```python
import os
import json
import pandas as pd

# ===============================
# PATHS
# ===============================
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), ".."))

INPUT_PATH = os.path.join(BASE_DIR, "data", "input", "new_dataset.csv")
CLEANED_PATH = os.path.join(BASE_DIR, "data", "cleaned", "df_cleaned.csv")
EVAL_PATH = os.path.join(BASE_DIR, "reports", "evaluation_reports", "evaluation_report.csv")

CONFIG_PATH = os.path.join(BASE_DIR, "notebooks", "config", "runtime_config.json")


# ===============================
# VALIDATE INPUT
# ===============================
def validate_input():
    if not os.path.exists(INPUT_PATH):
        raise FileNotFoundError(f"Input dataset not found at {INPUT_PATH}")

    df = pd.read_csv(INPUT_PATH)

    if df.empty:
        raise ValueError("Input dataset is empty")

    print("✅ Input validated")


# ===============================
# WRITE CONFIG (SHARED PATH)
# ===============================
def write_runtime_config():
    config = {
        "input_path": INPUT_PATH
    }

    os.makedirs(os.path.dirname(CONFIG_PATH), exist_ok=True)

    with open(CONFIG_PATH, "w") as f:
        json.dump(config, f, indent=4)

    print("✅ Runtime config written")


# ===============================
# VALIDATE OUTPUTS
# ===============================
def validate_outputs():
    print("\n[VALIDATION DEBUG]")
    print("Checking CLEANED_PATH:", CLEANED_PATH)
    print("Exists:", os.path.exists(CLEANED_PATH))

    print("Checking EVAL_PATH:", EVAL_PATH)
    print("Exists:", os.path.exists(EVAL_PATH))

    if not os.path.exists(CLEANED_PATH):
        raise FileNotFoundError(f"Cleaned dataset missing at {CLEANED_PATH}")

    if not os.path.exists(EVAL_PATH):
        raise FileNotFoundError(f"Evaluation report missing at {EVAL_PATH}")

    print("✅ Outputs validated")


# ===============================
# GET QUALITY SCORE
# ===============================
def get_quality_score():
    df = pd.read_csv(EVAL_PATH)

    if "quality_score" not in df.columns:
        raise ValueError("quality_score not found in evaluation report")

    return df["quality_score"].iloc[0]
```

```python
import os
import subprocess
import time
import json
import pandas as pd

INPUT_PATH = r"/media/ausaafkhan/New Volume1/project_exhibition/broken_dataset_repair/data/adult.csv"
BASE_DIR = os.getcwd()
if os.path.basename(BASE_DIR) == "notebooks":
    BASE_DIR = os.path.abspath(os.path.join(BASE_DIR, ".."))

CLEANED_PATH = os.path.join(BASE_DIR, "data", "cleaned", "df_cleaned.csv")

EVAL_PATH = os.path.join(BASE_DIR, "reports", "evaluation_reports", "evaluation_report.csv")
AUDIT_PATH = os.path.join(BASE_DIR, "reports", "audit_logs", "audit_log.csv")

LOG_PATH = os.path.join(BASE_DIR, "reports", "pipeline_reports", "pipeline_execution_log.json")
```

```python
import json
import os

CONFIG_PATH = "config/runtime_config.json"

def write_runtime_config():
    os.makedirs("config", exist_ok=True)

    config = {
        "input_path": INPUT_PATH
    }

    with open(CONFIG_PATH, "w") as f:
        json.dump(config, f, indent=4)
```

```python
def validate_input():
    if not os.path.exists(INPUT_PATH):
        raise FileNotFoundError(f"Missing input dataset: {INPUT_PATH}")

    if os.path.getsize(INPUT_PATH) == 0:
        raise ValueError("Input dataset is empty")

    try:
        pd.read_csv(INPUT_PATH, nrows=5)
    except Exception as e:
        raise ValueError(f"Input dataset unreadable: {str(e)}")
```

```python
import subprocess
import time
import os

def execute_step(step_name):
    start = time.time()

    # ---------------------------------
    # RESOLVE BASE DIR (ROBUST)
    # ---------------------------------
    BASE_DIR = os.getcwd()
    if os.path.basename(BASE_DIR) == "notebooks":
        BASE_DIR = os.path.abspath(os.path.join(BASE_DIR, ".."))

    input_nb = os.path.join(BASE_DIR, "notebooks", f"{step_name}.ipynb")
    output_nb = os.path.join(BASE_DIR, "notebooks", f"_executed_{step_name}.ipynb")

    # ---------------------------------
    # DEBUG (KEEP THIS FOR NOW)
    # ---------------------------------
    print("\n[DEBUG]")
    print("Current working dir:", os.getcwd())
    print("BASE_DIR:", BASE_DIR)
    print("Notebook path:", input_nb)
    print("Exists:", os.path.exists(input_nb))

    # ---------------------------------
    # VALIDATE NOTEBOOK EXISTS
    # ---------------------------------
    if not os.path.exists(input_nb):
        return {
            "step": step_name,
            "status": "FAILED",
            "execution_time": 0,
            "error": f"Missing notebook: {input_nb}"
        }

    try:
        # ---------------------------------
        # EXECUTE USING PAPERMILL
        # ---------------------------------
        result = subprocess.run(
            ["papermill", input_nb, output_nb],
            cwd=BASE_DIR,   # 🔥 critical fix
            capture_output=True,
            text=True
        )

        runtime = round(time.time() - start, 4)

        # ---------------------------------
        # RELIABLE FAILURE DETECTION
        # ---------------------------------
        error_text = result.stderr if result.stderr else ""

        if "PapermillExecutionError" in error_text:
            return {
                "step": step_name,
                "status": "FAILED",
                "execution_time": runtime,
                "error": error_text[-3000:]
            }

        # ---------------------------------
        # SUCCESS
        # ---------------------------------
        return {
            "step": step_name,
            "status": "SUCCESS",
            "execution_time": runtime,
            "error": None
        }

    except Exception as e:
        return {
            "step": step_name,
            "status": "FAILED",
            "execution_time": round(time.time() - start, 4),
            "error": str(e)
        }
```

```python
import json
import os

def save_log():
    os.makedirs(os.path.dirname(LOG_PATH), exist_ok=True)

    with open(LOG_PATH, "w") as f:
        json.dump(PIPELINE_LOG, f, indent=4)

    print(f"📊 Pipeline log saved to: {LOG_PATH}")
```

```python
PIPELINE_LOG = []

def run_pipeline():
    global PIPELINE_LOG
    PIPELINE_LOG = []

    for step in PIPELINE_STEPS:
        print(f"\n▶ Running: {step}")

        result = execute_step(step)

        # ✅ print result immediately
        print("Step Result:", result)

        PIPELINE_LOG.append(result)

        # ---------------------------------
        # STRICT FAIL-FAST
        # ---------------------------------
        if result["status"] != "SUCCESS":
            print("\n❌ FAILED at:", step)
            print("\n--- ERROR DETAILS ---")
            print(result["error"])
            print("\n--- END ERROR ---")

            save_log()
            raise RuntimeError(f"Pipeline stopped at {step}")

    # ---------------------------------
    # ALL SUCCESS
    # ---------------------------------
    save_log()
    print("\n✅ Pipeline execution completed")
```

```python
def validate_outputs():
    if not os.path.exists(CLEANED_PATH):
        raise FileNotFoundError("Cleaned dataset missing")

    if not os.path.exists(EVAL_PATH):
        raise FileNotFoundError("Evaluation report missing")

    if not os.path.exists(AUDIT_PATH):
        raise FileNotFoundError(f"Audit log missing at {AUDIT_PATH}")

    if not os.path.exists(LOG_PATH):
        raise FileNotFoundError(f"Pipeline log missing at {LOG_PATH}")

    print("✅ Logs validated")
```

```python
def get_quality_score():
    df = pd.read_csv(EVAL_PATH)

    if "quality_score" not in df.columns:
        raise ValueError("quality_score column missing")

    if df.empty:
        raise ValueError("evaluation report is empty")

    return float(df["quality_score"].iloc[0])
```

```python
if __name__ == "__main__":
    print("\n🚀 STARTING PIPELINE")

    validate_input()
    write_runtime_config()
    run_pipeline()
    validate_outputs()

    score = get_quality_score()

    print("\n🎯 FINAL QUALITY SCORE:", score)
```
