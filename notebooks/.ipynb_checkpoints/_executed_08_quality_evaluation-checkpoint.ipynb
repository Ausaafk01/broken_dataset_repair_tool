```python
import pandas as pd
import numpy as np
import os

NUMERIC_TOLERANCE = 1e-6
```

```python
def find_project_root(start_path):
    current = os.path.abspath(start_path)
    while True:
        contents = os.listdir(current)
        if "data" in contents and "reports" in contents:
            return current
        parent = os.path.dirname(current)
        if parent == current:
            raise FileNotFoundError("Project root not found")
        current = parent

project_root = find_project_root(os.getcwd())

clean_path = os.path.join(project_root, "data/clean/clean_dataset.csv")
cleaned_path = os.path.join(project_root, "data/cleaned/df_cleaned.csv")
audit_log_path = os.path.join(project_root, "reports/audit_logs/audit_log.csv")

for path in [clean_path, cleaned_path, audit_log_path]:
    if not os.path.exists(path):
        raise FileNotFoundError(f"Required file missing: {path}")

print("Project root:", project_root)
```

```python
df_clean = pd.read_csv(clean_path)
df_cleaned = pd.read_csv(cleaned_path)
audit_log = pd.read_csv(audit_log_path)
```

```python
if df_clean.empty:
    raise ValueError("df_clean is empty")

if df_cleaned.empty:
    raise ValueError("df_cleaned is empty")

if audit_log.empty:
    raise ValueError("audit_log is empty")
```

```python
for column in df_clean.columns:

    if column not in df_cleaned.columns:
        continue

    clean_dtype = df_clean[column].dtype
    cleaned_dtype = df_cleaned[column].dtype

    # Detect encoding (categorical → numeric)
    if pd.api.types.is_object_dtype(clean_dtype) and pd.api.types.is_numeric_dtype(cleaned_dtype):
        raise ValueError(f"Encoding detected in column: {column}")

    # Detect numeric → object corruption
    if pd.api.types.is_numeric_dtype(clean_dtype) and pd.api.types.is_object_dtype(cleaned_dtype):
        raise ValueError(f"Unexpected type conversion in column: {column}")
```

```python
if "id" not in df_clean.columns or "id" not in df_cleaned.columns:
    raise ValueError("Primary key 'id' missing")

if not df_clean["id"].is_unique:
    raise ValueError("df_clean contains duplicate IDs")

if not df_cleaned["id"].is_unique:
    raise ValueError("df_cleaned contains duplicate IDs")

df_clean_indexed = df_clean.set_index("id")
df_cleaned_indexed = df_cleaned.set_index("id")

common_ids = sorted(df_clean_indexed.index.intersection(df_cleaned_indexed.index))

if len(common_ids) == 0:
    raise ValueError("No overlapping IDs")

df_clean_aligned = df_clean_indexed.loc[common_ids]
df_cleaned_aligned = df_cleaned_indexed.loc[common_ids]

if not df_clean_aligned.index.equals(df_cleaned_aligned.index):
    raise ValueError("Index alignment failed")
```

```python
evaluation_rows = []

for column in df_clean_aligned.columns:

    if column not in df_cleaned_aligned.columns:
        total_values = len(df_clean_aligned)
        matching_values = 0
        accuracy = 0.0
    else:
        clean_series = df_clean_aligned[column]
        cleaned_series = df_cleaned_aligned[column]

        total_values = len(clean_series)

        if pd.api.types.is_numeric_dtype(clean_series):
            diff = np.abs(clean_series - cleaned_series)
            matches = diff <= NUMERIC_TOLERANCE
        else:
            matches = clean_series == cleaned_series

        matching_values = int(matches.sum())
        accuracy = matching_values / total_values if total_values > 0 else 0.0

    evaluation_rows.append({
        "column_name": column,
        "total_values": total_values,
        "matching_values": matching_values,
        "accuracy": accuracy
    })
```

```python
evaluation_report = pd.DataFrame(evaluation_rows)

if evaluation_report.empty:
    raise ValueError("evaluation_report is empty")
```

```python
quality_score = float(evaluation_report["accuracy"].mean())
```

```python
expected_columns = ["column_name", "total_values", "matching_values", "accuracy"]

if list(evaluation_report.columns) != expected_columns:
    raise ValueError("Schema mismatch")

if evaluation_report.isnull().any().any():
    raise ValueError("Null values detected")
```

```python
output_dir = os.path.join(project_root, "reports/evaluation_reports")
output_file = os.path.join(output_dir, "evaluation_report.csv")

os.makedirs(output_dir, exist_ok=True)

evaluation_report.to_csv(output_file, index=False)

print("Saved to:", output_file)
```

```python
print("Evaluation complete.")
print(f"Quality Score: {quality_score:.6f}")
```
