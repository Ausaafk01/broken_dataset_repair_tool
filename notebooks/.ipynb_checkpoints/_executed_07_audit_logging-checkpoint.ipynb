```python
# ============================================
# CELL 1 — IMPORTS
# ============================================

import pandas as pd
import numpy as np
import os
```

```python
# ============================================
# CELL 2 — LOAD DATA (STRICT FROM DISK)
# ============================================

def resolve_path(path):
    if os.path.exists(path):
        return path
    alt = os.path.join("..", path)
    if os.path.exists(alt):
        return alt
    raise FileNotFoundError(f"File not found: {path}")

df_working = pd.read_csv(resolve_path("data/real/df_working.csv"))
df_cleaned = pd.read_csv(resolve_path("data/cleaned/df_cleaned.csv"))
repair_plan = pd.read_csv(resolve_path("reports/repair_plan/repair_plan.csv"))
```

```python
# ============================================
# CELL 3 — VALIDATION (FAIL-FAST)
# ============================================

if df_working.empty:
    raise ValueError("df_working is empty")

if df_cleaned.empty:
    raise ValueError("df_cleaned is empty")

if repair_plan.empty:
    raise ValueError("repair_plan is empty")
```

```python
# ============================================
# CELL 4 — ACTION TO CHANGE TYPE (STRICT)
# ============================================

ACTION_TO_CHANGE_TYPE = {
    "impute_missing": "value_imputed",
    "correct_invalid": "value_corrected",
    "cap_outliers": "value_capped",
    "evaluate_high_cardinality": "no_change",
    "mark_as_identifier": "no_change",
    "resolve_semantic_mismatch": "no_change",
    "flag_for_review": "no_change"
}
```

```python
# ============================================
# CELL 5 — INITIALIZE AUDIT LOG
# ============================================

audit_records = []
```

```python
# ============================================
# CELL 6 — AUDIT LOOP (DETERMINISTIC + NO ENCODING CHECK)
# ============================================

for _, row in repair_plan.iterrows():
    column_name = row["column_name"]
    action = row["recommended_action"]

    if action not in ACTION_TO_CHANGE_TYPE:
        raise ValueError(f"Unsupported action: {action}")

    change_type = ACTION_TO_CHANGE_TYPE[action]

    if column_name not in df_working.columns:
        raise ValueError(f"Column missing in df_working: {column_name}")

    working_col = df_working[column_name]

    if column_name not in df_cleaned.columns:
        raise ValueError(f"Column missing in df_cleaned: {column_name}")

    cleaned_col = df_cleaned[column_name]

    # -----------------------------
    # REPRESENTATION CONSISTENCY CHECK
    # -----------------------------

    # Check dtype mismatch that implies encoding
    if working_col.dtype == object and np.issubdtype(cleaned_col.dtype, np.number):
        raise ValueError(f"Encoding detected in column: {column_name}")

    # Check string-to-numeric mapping
    if working_col.dtype == object:
        working_unique = set(working_col.dropna().unique())
        cleaned_unique = set(cleaned_col.dropna().unique())

        # If cleaned values are numeric while original were strings
        if all(isinstance(x, str) for x in working_unique):
            if all(isinstance(x, (int, float, np.number)) for x in cleaned_unique):
                raise ValueError(f"Categorical encoding detected in column: {column_name}")

    # -----------------------------
    # VALUE COMPARISON
    # -----------------------------

    values_equal = working_col.equals(cleaned_col)
    was_modified = not values_equal

    # -----------------------------
    # DETAILS (DETERMINISTIC)
    # -----------------------------

    if change_type == "value_imputed":
        missing_before = working_col.isna().sum()
        missing_after = cleaned_col.isna().sum()
        filled = missing_before - missing_after
        details = f"{filled} values imputed"

    elif change_type == "value_corrected":
        changes = (working_col != cleaned_col).sum()
        details = f"{changes} values corrected"

    elif change_type == "value_capped":
        changes = (working_col != cleaned_col).sum()
        details = f"{changes} values capped"

    else:
        details = "no transformation applied"

    audit_records.append({
        "column_name": column_name,
        "action_applied": action,
        "was_modified": was_modified,
        "change_type": change_type,
        "details": details
    })
```

```python
# ============================================
# CELL 7 — CREATE AUDIT LOG DATAFRAME
# ============================================

audit_log = pd.DataFrame(audit_records)
```

```python
# ============================================
# CELL 8 — FINAL VALIDATION (FAIL-FAST)
# ============================================

required_columns = [
    "column_name",
    "action_applied",
    "was_modified",
    "change_type",
    "details"
]

if audit_log.empty:
    raise ValueError("audit_log is empty")

if list(audit_log.columns) != required_columns:
    raise ValueError("audit_log schema mismatch")

if audit_log.isnull().any().any():
    raise ValueError("audit_log contains null values")
```

```python
# ============================================
# CELL 9 — SAVE OUTPUT (FIXED)
# ============================================

import os

# Step 1 — Resolve project root
def find_project_root(start_path):
    current = os.path.abspath(start_path)
    while True:
        contents = os.listdir(current)
        if "data" in contents and "reports" in contents:
            return current
        parent = os.path.dirname(current)
        if parent == current:
            raise FileNotFoundError("Project root not found")
        current = parent

project_root = find_project_root(os.getcwd())

# Step 2 — Define directory and file path separately
output_dir = os.path.join(project_root, "reports/audit_logs")
output_file = os.path.join(output_dir, "audit_log.csv")

# Step 3 — Ensure directory exists
os.makedirs(output_dir, exist_ok=True)

# Step 4 — Save file
audit_log.to_csv(output_file, index=False)

print("Audit log saved to:", output_file)
```
