```python
import os
import pandas as pd
# ============================================================
# ABSOLUTE PROJECT ROOT RESOLUTION (FINAL FIX)
# ============================================================

import os

def find_project_root():
    current = os.getcwd()

    while True:
        if os.path.basename(current) == "broken_dataset_repair":
            return current

        parent = os.path.dirname(current)

        if parent == current:
            raise RuntimeError("❌ Could not locate 'broken_dataset_repair' root")

        current = parent


BASE_DIR = find_project_root()

print("✅ BASE_DIR:", BASE_DIR)

INPUT_PATH = os.path.join(PROJECT_ROOT, "reports/issue_reports/detected_issues.csv")
OUTPUT_PATH = os.path.join(PROJECT_ROOT, "reports/repair_plan/repair_plan.csv")
```

```python
if not os.path.exists(INPUT_PATH):
    raise FileNotFoundError(f"Input file not found: {INPUT_PATH}")

detected_issues = pd.read_csv(INPUT_PATH)

if detected_issues.empty:
    raise ValueError("detected_issues is empty")

required_input_cols = ["column_name", "issue_type"]
missing_cols = [c for c in required_input_cols if c not in detected_issues.columns]

if missing_cols:
    raise ValueError(f"detected_issues missing required columns: {missing_cols}")

print("✅ detected_issues loaded and validated")
print(detected_issues.head())
```

```python
def map_issue(issue_type):
    if issue_type == "missing_values":
        return "impute_missing", "medium", 0.7, "Missing values detected"
    
    elif issue_type == "high_cardinality":
        return "evaluate_high_cardinality", "medium", 0.7, "High cardinality column"
    
    elif issue_type == "id_column":
        return "mark_as_identifier", "low", 0.5, "Column appears to be identifier"
    
    else:
        raise ValueError(f"Unknown issue_type encountered: {issue_type}")
```

```python
records = []

for _, row in detected_issues.iterrows():
    col = row["column_name"]
    issue = row["issue_type"]

    action, priority, confidence, details = map_issue(issue)

    records.append({
        "column_name": col,
        "issue_type": issue,
        "recommended_action": action,
        "action_priority": priority,
        "confidence": confidence,
        "details": details
    })

repair_plan = pd.DataFrame(records)
```

```python
if repair_plan.empty:
    raise ValueError("repair_plan is empty")

if len(repair_plan) != len(detected_issues):
    raise ValueError("Row count mismatch between detected_issues and repair_plan")

if repair_plan.isnull().any().any():
    raise ValueError("repair_plan contains null values")

REQUIRED_OUTPUT_COLUMNS = [
    "column_name",
    "issue_type",
    "recommended_action",
    "action_priority",
    "confidence",
    "details"
]

if list(repair_plan.columns) != REQUIRED_OUTPUT_COLUMNS:
    raise ValueError("repair_plan schema mismatch")

# Deterministic sorting
repair_plan = repair_plan.sort_values(by=["column_name"]).reset_index(drop=True)

print("✅ repair_plan validated")
```

```python
print("\n===== REPAIR PLAN AUDIT =====")

print(f"Total actions: {len(repair_plan)}")

print("\nActions per issue_type:")
print(repair_plan["issue_type"].value_counts())

print("\nActions per priority:")
print(repair_plan["action_priority"].value_counts())
```

```python
os.makedirs(os.path.dirname(OUTPUT_PATH), exist_ok=True)

repair_plan.to_csv(OUTPUT_PATH, index=False)

print(f"\n✅ repair_plan saved to: {OUTPUT_PATH}")
```
