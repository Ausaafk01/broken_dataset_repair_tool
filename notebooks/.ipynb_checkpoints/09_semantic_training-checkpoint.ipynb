```python
import os
import json
import joblib
import pandas as pd
import numpy as np
import re

from sklearn.ensemble import RandomForestClassifier

# Paths
BASE_PATH = "/media/ausaafkhan/New Volume1/Project exhibition/broken_dataset_repair"

PROFILE_PATH = f"{BASE_PATH}/reports/profiling_reports/dataset_profile.csv"
CLEAN_DATA_PATH = f"{BASE_PATH}/data/cleaned/clean_dataset.csv"
TRAINING_DATA_PATH = f"{BASE_PATH}/models/semantic_training_data.csv"
MODEL_PATH = f"{BASE_PATH}/models/semantic_model.pkl"
REPORT_PATH = f"{BASE_PATH}/reports/semantic_reports/semantic_model_report.json"

os.makedirs(os.path.dirname(MODEL_PATH), exist_ok=True)
os.makedirs(os.path.dirname(REPORT_PATH), exist_ok=True)
```

```python
if not os.path.exists(PROFILE_PATH):
    raise FileNotFoundError("dataset_profile.csv not found")

df_profile = pd.read_csv(PROFILE_PATH)

df_clean = None
if os.path.exists(CLEAN_DATA_PATH):
    df_clean = pd.read_csv(CLEAN_DATA_PATH)

df_training = None
if os.path.exists(TRAINING_DATA_PATH):
    df_training = pd.read_csv(TRAINING_DATA_PATH)

print("Inputs Loaded:")
print(f"Profile: {df_profile.shape}")
print(f"Clean dataset: {None if df_clean is None else df_clean.shape}")
print(f"Training data: {None if df_training is None else df_training.shape}")
```

```python
def extract_features(series: pd.Series):
    total = len(series)

    if total == 0:
        return None

    series_str = series.astype(str)

    num_numeric = series_str.str.match(r"^-?\d+(\.\d+)?$").sum()
    num_alpha = series_str.str.match(r"^[A-Za-z]+$").sum()
    num_missing = series.isna().sum()

    avg_length = series_str.str.len().mean()
    nunique = series.nunique(dropna=True)

    has_email = series_str.str.contains(r"@", regex=True).any()
    has_date = series_str.str.contains(r"\d{4}-\d{2}-\d{2}|\d{2}/\d{2}/\d{4}", regex=True).any()

    numeric_values = pd.to_numeric(series, errors='coerce')
    has_numeric_range = numeric_values.notna().sum() > 0

    return {
        "pct_numeric": num_numeric / total,
        "pct_alpha": num_alpha / total,
        "pct_missing": num_missing / total,
        "avg_length": avg_length,
        "nunique": nunique,
        "has_email": int(has_email),
        "has_date": int(has_date),
        "has_numeric_range": int(has_numeric_range)
    }
```

```python
# 🔒 SAFE LABELS — YOU DEFINE THIS
# MUST match real known schema (edit as needed)

MANUAL_LABELS = {
    "name": "name",
    "age": "age",
    "gender": "gender",
    "email": "email",
    "dob": "date",
    "salary": "salary"
}

if df_clean is None:
    raise ValueError("Clean dataset required for safe labeling")

labeled_rows = []

for col in df_clean.columns:
    if col not in MANUAL_LABELS:
        continue

    features = extract_features(df_clean[col])
    if features is None:
        continue

    features["label"] = MANUAL_LABELS[col]
    features["column_name"] = col

    labeled_rows.append(features)

df_new_training = pd.DataFrame(labeled_rows)

if df_new_training.empty:
    raise ValueError("No labeled training data generated")

print("New training samples:", df_new_training.shape)
```

```python
if df_training is not None:
    df_combined = pd.concat([df_training, df_new_training], ignore_index=True)
else:
    df_combined = df_new_training.copy()

# Deterministic deduplication
df_combined = df_combined.drop_duplicates(
    subset=[
        "pct_numeric",
        "pct_alpha",
        "pct_missing",
        "avg_length",
        "nunique",
        "has_email",
        "has_date",
        "has_numeric_range",
        "label"
    ]
).reset_index(drop=True)

if df_combined.empty:
    raise ValueError("Training dataset is empty after deduplication")

print("Final training dataset:", df_combined.shape)
```

```python
FEATURE_COLS = [
    "pct_numeric",
    "pct_alpha",
    "pct_missing",
    "avg_length",
    "nunique",
    "has_email",
    "has_date",
    "has_numeric_range"
]

X = df_combined[FEATURE_COLS]
y = df_combined["label"]

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=1  # ensures deterministic behavior
)

model.fit(X, y)

print("Model trained successfully")
```

```python
joblib.dump(model, MODEL_PATH)
df_combined.to_csv(TRAINING_DATA_PATH, index=False)

print("Model and training data saved")
```

```python
report = {
    "num_training_samples": int(len(df_combined)),
    "num_features": len(FEATURE_COLS),
    "labels": sorted(df_combined["label"].unique().tolist()),
    "model_type": "RandomForestClassifier",
    "deterministic": True
}

with open(REPORT_PATH, "w") as f:
    json.dump(report, f, indent=4)

print("Training report saved")
```
