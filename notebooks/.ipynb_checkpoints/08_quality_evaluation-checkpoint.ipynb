```python
import pandas as pd
import numpy as np
import json
import os
from pathlib import Path
```

```python
BASE_PATH = "../"   # go from notebooks → project root

CLEANED_PATH = BASE_PATH + "data/cleaned/df_cleaned.csv"
GROUND_TRUTH_DIR = BASE_PATH + "data/cleaned/"
PROFILE_DIR = BASE_PATH + "reports/profiling_reports/"
SEMANTIC_TYPES_PATH = BASE_PATH + "reports/semantic_types.json"
AUDIT_LOG_PATH = BASE_PATH + "reports/audit_logs/audit_log.json"

# Load cleaned dataset
df_cleaned = pd.read_csv(CLEANED_PATH)

# Load ground truth (assumes single file)
ground_truth_files = list(Path(GROUND_TRUTH_DIR).glob("*.csv"))
assert len(ground_truth_files) == 1, "Expected exactly one ground truth file"
df_truth = pd.read_csv(ground_truth_files[0])

# Load semantic types
with open(SEMANTIC_TYPES_PATH, "r") as f:
    semantic_types = json.load(f)

# Load audit log
with open(AUDIT_LOG_PATH, "r") as f:
    audit_log = json.load(f)

# Align indices
df_cleaned = df_cleaned.reset_index(drop=True)
df_truth = df_truth.reset_index(drop=True)

assert df_cleaned.shape == df_truth.shape, "Shape mismatch between cleaned and ground truth"
```

```python
def get_column_type(col):
    return semantic_types.get(col, "unknown")

def safe_compare(a, b):
    return (a == b) | (pd.isna(a) & pd.isna(b))
```

```python
def evaluate_column_accuracy(df_cleaned, df_truth):
    results = {}

    for col in df_cleaned.columns:
        col_type = get_column_type(col)

        cleaned = df_cleaned[col]
        truth = df_truth[col]

        match_mask = safe_compare(cleaned, truth)
        exact_match_ratio = match_mask.mean()

        col_result = {
            "exact_match_ratio": float(exact_match_ratio)
        }

        if col_type == "numeric":
            mae = np.mean(np.abs(cleaned - truth))
            col_result["mae"] = float(mae)

        elif col_type in ["categorical", "boolean"]:
            col_result["category_accuracy"] = float(exact_match_ratio)

        elif col_type in ["datetime", "date"]:
            col_result["date_accuracy"] = float(exact_match_ratio)

        results[col] = col_result

    return results
```

```python
def evaluate_missing_recovery(df_cleaned, df_truth):
    total_missing = 0
    correctly_restored = 0

    for col in df_cleaned.columns:
        truth_missing = df_truth[col].isna()
        cleaned_missing = df_cleaned[col].isna()

        total_missing += truth_missing.sum()

        restored = (~cleaned_missing) & truth_missing
        correct_restored = restored & safe_compare(df_cleaned[col], df_truth[col])

        correctly_restored += correct_restored.sum()

    return float(correctly_restored / total_missing) if total_missing > 0 else 1.0
```

```python
def evaluate_type_correctness(df_cleaned, df_truth):
    matches = []

    for col in df_cleaned.columns:
        matches.append(df_cleaned[col].dtype == df_truth[col].dtype)

    return float(np.mean(matches))
```

```python
def evaluate_outliers(df_cleaned, df_truth):
    total_changed = 0
    correct_fixes = 0

    for col in df_cleaned.columns:
        if get_column_type(col) != "numeric":
            continue

        cleaned = df_cleaned[col]
        truth = df_truth[col]

        changed = ~safe_compare(cleaned, truth)

        total_changed += changed.sum()
        correct_fixes += (changed & safe_compare(cleaned, truth)).sum()

    if total_changed == 0:
        return 1.0

    return float(correct_fixes / total_changed)
```

```python
def evaluate_duplicates(df_cleaned, df_truth):
    cleaned_dupes = df_cleaned.duplicated().sum()
    truth_dupes = df_truth.duplicated().sum()

    if truth_dupes == 0:
        return 1.0 if cleaned_dupes == 0 else 0.0

    return float(1 - abs(cleaned_dupes - truth_dupes) / truth_dupes)
```

```python
def evaluate_overcorrection(df_cleaned, df_truth):
    total_compared = 0
    unnecessary_changes = 0

    for col in df_cleaned.columns:
        cleaned = df_cleaned[col]
        truth = df_truth[col]

        mismatch = ~safe_compare(cleaned, truth)

        total_compared += len(cleaned)
        unnecessary_changes += mismatch.sum()

    return float(unnecessary_changes / total_compared)
```

```python
def evaluate_audit_consistency(df_cleaned, df_truth, audit_log):
    actual_changes = 0

    for col in df_cleaned.columns:
        actual_changes += (~safe_compare(df_cleaned[col], df_truth[col])).sum()

    logged_changes = sum(entry.get("rows_affected", 0) for entry in audit_log)

    if actual_changes == 0:
        return 1.0 if logged_changes == 0 else 0.0

    diff = abs(actual_changes - logged_changes)

    return float(1 - (diff / actual_changes))
```

```python
def compute_final_score(column_scores,
                        missing_score,
                        type_score,
                        outlier_score,
                        duplicate_score,
                        overcorrection_rate,
                        audit_score):

    weights = {
        "column": 0.4,
        "missing": 0.15,
        "type": 0.1,
        "outlier": 0.1,
        "duplicate": 0.1,
        "audit": 0.15
    }

    avg_column_score = np.mean([
        col["exact_match_ratio"] for col in column_scores.values()
    ])

    base_score = (
        weights["column"] * avg_column_score +
        weights["missing"] * missing_score +
        weights["type"] * type_score +
        weights["outlier"] * outlier_score +
        weights["duplicate"] * duplicate_score +
        weights["audit"] * audit_score
    )

    penalty = overcorrection_rate * 0.5

    final_score = max(0.0, base_score - penalty)

    return float(final_score), float(penalty)
```

```python
column_scores = evaluate_column_accuracy(df_cleaned, df_truth)
missing_score = evaluate_missing_recovery(df_cleaned, df_truth)
type_score = evaluate_type_correctness(df_cleaned, df_truth)
outlier_score = evaluate_outliers(df_cleaned, df_truth)
duplicate_score = evaluate_duplicates(df_cleaned, df_truth)
overcorrection_rate = evaluate_overcorrection(df_cleaned, df_truth)
audit_score = evaluate_audit_consistency(df_cleaned, df_truth, audit_log)

overall_score, penalty = compute_final_score(
    column_scores,
    missing_score,
    type_score,
    outlier_score,
    duplicate_score,
    overcorrection_rate,
    audit_score
)

quality_score = {
    "overall_score": overall_score,
    "column_scores": column_scores,
    "missing_recovery_score": missing_score,
    "type_correctness_score": type_score,
    "outlier_score": outlier_score,
    "duplicate_score": duplicate_score,
    "overcorrection_penalty": penalty,
    "audit_consistency_score": audit_score
}

quality_score
```
