```python
# ============================================================
# NOTEBOOK 04: ISSUE DETECTION (STRICT)
# ============================================================

import os
import pandas as pd

# ============================================================
# PROJECT ROOT (LOCKED)
# ============================================================

PROJECT_ROOT = "/media/ausaafkhan/New Volume1/Project exhibition/broken_dataset_repair"

PROFILE_PATH = os.path.join(PROJECT_ROOT, "reports/profiling_reports/dataset_profile.csv")

OUTPUT_DIR = os.path.join(PROJECT_ROOT, "reports/issue_reports")
OUTPUT_PATH = os.path.join(OUTPUT_DIR, "detected_issues.csv")

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================================
# FAIL-FAST INPUT CHECK
# ============================================================

if not os.path.exists(PROFILE_PATH):
    raise FileNotFoundError(f"Missing dataset_profile: {PROFILE_PATH}")
```

```python
# ============================================================
# LOAD DATA
# ============================================================

dataset_profile = pd.read_csv(PROFILE_PATH)

# ============================================================
# BASE VALIDATION
# ============================================================

if dataset_profile.empty:
    raise ValueError("dataset_profile is empty")

base_required = [
    "column_name",
    "null_ratio",
    "unique_ratio",
    "dtype"
]

missing_base = [col for col in base_required if col not in dataset_profile.columns]

if missing_base:
    raise ValueError(f"Missing required base columns: {missing_base}")

# ============================================================
# DERIVE n_unique (DETERMINISTIC)
# ============================================================

if "n_unique" not in dataset_profile.columns:
    if "unique_count" in dataset_profile.columns:
        dataset_profile["n_unique"] = dataset_profile["unique_count"]
    elif "row_count" in dataset_profile.columns:
        dataset_profile["n_unique"] = (
            dataset_profile["unique_ratio"] * dataset_profile["row_count"]
        ).astype(int)
    else:
        raise ValueError("Cannot derive n_unique — missing unique_count or row_count")

# ============================================================
# FINAL VALIDATION
# ============================================================

required_cols = [
    "column_name",
    "null_ratio",
    "unique_ratio",
    "n_unique",
    "dtype"
]

missing = [col for col in required_cols if col not in dataset_profile.columns]

if missing:
    raise ValueError(f"Missing required columns after derivation: {missing}")

print("✅ dataset_profile validated")
```

```python
# ============================================================
# ISSUE DETECTION ENGINE (FINAL — WORKING)
# ============================================================

issues = []

def add_issue(col, issue):
    issues.append({
        "column_name": col,
        "issue_type": issue
    })

# ============================================================
# DETECTION LOOP
# ============================================================

for _, row in dataset_profile.iterrows():
    col = str(row["column_name"])
    col_lower = col.lower()

    null_ratio = float(row["null_ratio"])
    unique_ratio = float(row["unique_ratio"])
    n_unique = int(row["n_unique"])
    dtype = str(row["dtype"])

    added_issues = set()

    # --------------------------------------------------------
    # 1. missing_values (explicit)
    # --------------------------------------------------------
    if null_ratio > 0:
        add_issue(col, "missing_values")
        added_issues.add("missing_values")

    # --------------------------------------------------------
    # 2. implicit missing_values (FIXED USING n_unique)
    # --------------------------------------------------------
    if (
        dtype in ["object", "str"]
        and n_unique > 2
        and n_unique < 100
        and "missing_values" not in added_issues
    ):
        add_issue(col, "missing_values")
        added_issues.add("missing_values")

    # --------------------------------------------------------
    # 3. high_cardinality
    # --------------------------------------------------------
    if unique_ratio > 0.9:
        add_issue(col, "high_cardinality")

    # --------------------------------------------------------
    # 4. id_column
    # --------------------------------------------------------
    if "id" in col_lower:
        add_issue(col, "id_column")
```

```python
# ============================================================
# BUILD OUTPUT
# ============================================================

detected_issues = pd.DataFrame(issues)

# ------------------------------------------------------------
# FAIL-FAST
# ------------------------------------------------------------
if detected_issues.empty:
    raise ValueError("No issues detected")

# ------------------------------------------------------------
# STRICT SCHEMA
# ------------------------------------------------------------
detected_issues = detected_issues[["column_name", "issue_type"]]

# ------------------------------------------------------------
# NO NULL VALUES
# ------------------------------------------------------------
if detected_issues.isnull().any().any():
    raise ValueError("Null values found in detected_issues")

# ------------------------------------------------------------
# REMOVE DUPLICATES
# ------------------------------------------------------------
detected_issues = detected_issues.drop_duplicates()

# ------------------------------------------------------------
# DETERMINISTIC SORT
# ------------------------------------------------------------
detected_issues = detected_issues.sort_values(
    by=["column_name", "issue_type"]
).reset_index(drop=True)

print("✅ detected_issues built successfully")
```

```python
# ============================================================
# SAVE OUTPUT
# ============================================================

detected_issues.to_csv(OUTPUT_PATH, index=False)

print("✅ detected_issues.csv saved")
print(OUTPUT_PATH)
```
